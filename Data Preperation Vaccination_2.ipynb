# Import Vaccination Data

Github:https://github.com/govex/COVID-19/blob/master/data_tables/vaccine_data/global_data/time_series_covid19_vaccine_global.csv

```python
vcc_data = r"C:\Users\Tyson\OneDrive\Desktop\Data Enterprise\My_Data\covid_19_analysis_3_countries\data\raw\COVID-19\time_series_covid19_vaccine_global.csv"
vcc_data= pd.read_csv(vcc_data)
```

```python
vcc_data.head()
```

```python
#Renaming and deleting unwanted columns
df1= vcc_data.drop(['Doses_admin', 'People_partially_vaccinated', 'Report_Date_String', 'UID', 'Province_State'], axis=1, inplace=True)
df2=vcc_data.rename(columns={'Country_Region': 'Country', 'People_fully_vaccinated': 'Vaccindated_People'})
```

```python
df2.head()
```

# Vaccinated people per population of three countries

```python
#Italy's vaccinated people per population
df_Italy = df2[df2["Country"] == 'Italy'].copy()
df_Italy['Vaccindated_People'] = df_Italy['Vaccindated_People'].div(59550000)
```

```python
#Kazakhstan's vaccinated people per population
df_Kazakhstan = df2[df2["Country"]=='Kazakhstan'].copy()
df_Kazakhstan['Vaccindated_People'] = df_Kazakhstan['Vaccindated_People'].div(18750000)
```

```python
#Armenia's vaccinated people per population
df_Armenia = df2[df2["Country"]=='Armenia'].copy()
df_Armenia['Vaccindated_People'] = df_Armenia['Vaccindated_People'].div(2963000)
```

```python
#Deleting the NaN values of three countries
italy= df_Italy.dropna()
kazakhstan = df_Kazakhstan.dropna()
armenia = df_Armenia.dropna()
```

```python
#creating a list of the three countries
df_country = [italy, kazakhstan, armenia]
```

```python
#Merging three countries to a single dataframe using concat method
result_df = pd.concat(df_country).reset_index(drop=True)
```

```python
#final dataframe
result_df
```

```python
#Converting in to date format
result_df['Date'] =  pd.to_datetime(result_df['Date'], format='%Y-%m-%d')
```

```python
#Plot the Graph of the three countries 
plot = pe.line(result_df, x="Date", y="Vaccindated_People", color = 'Country', title='Vaccination Rate per population of Italy, Kazhakhstan and, Armenia')
plot.show()
```
